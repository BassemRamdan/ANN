```python
import numpy as np
```

```python
def MeanSquareErorr(y_pred,y_test):
    sub=np.subtract(y_test,y_pred)
    MeanSquareErorr=np.mean((sub)**2)
    return MeanSquareErorr
```

```python
y_pred=np.array([1,2,3,4,5,6])
y_test=np.array([1,3,3,5,5,6])

MeanSquareErorr(y_pred,y_test)
```

```python
def Sigmoid(x):
    return 1/(1+np.exp(-x))


z=np.arange(1,10,2).astype(int)
print("Sigmoid",Sigmoid(z))
```

```python
def Relu(x):

    return np.maximum(0,x)

z=np.arange(1,20,2)
print('Relu',Relu(z))
```
